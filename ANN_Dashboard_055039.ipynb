  **4.1 Model Hyperparameters**
  These hyperparameters are settings that are configured before the training process begins and significantly influence how the model learns and performs.
The breakdown of hyperparameters for the given model -
1. **Epochs: 25**
  
 The model was trained for 25 epochs. An epoch represents one complete pass through the entire training dataset.
2. **Learning Rate: 0.001**

 The learning rate controls the step size at each iteration while the model's weights are being updated during training (using the optimizer).
3. **Activation Function: relu (Rectified Linear Unit)**

 ReLU is the activation function applied to the neurons in the hidden layers of the network. It introduces non-linearity to the model, allowing it to learn complex relationships in the data. ReLU outputs the input directly if it's positive, and zero otherwise.
4. **Optimizer: adam**

 Adam (Adaptive Moment Estimation) is an adaptive learning rate optimization algorithm. It combines the benefits of two other popular optimizers, AdaGrad and RMSProp. Adam adapts the learning rates for each parameter based on the first and second moments of the gradients.

5. **Number of Dense Layers: 2**

 The model has two fully connected (dense) hidden layers. Each neuron in a dense layer is connected to every neuron in the previous and subsequent layers.The number of layers determines the depth of the network and its capacity to learn complex features. Two dense layers suggest a relatively shallow network.
6. **Neurons per Layer: 32**

 Each of the two dense hidden layers contains 32 neurons. The number of neurons in a layer determines its width and representational capacity. More neurons allow the layer to learn more intricate patterns.32 neurons per layer represent a moderate width.
7. **Dropout Rate: 0.20**

 Dropout is a regularization technique where, during each training iteration, a randomly selected fraction (in this case, 20% or 0.20) of neurons in a layer are "dropped out" (i.e., their output is temporarily ignored). This prevents the network from relying too heavily on any single neuron and promotes more robust learning and better generalization to unseen data.
